```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv("/content/stock.csv")

plt.figure(figsize=(14, 6))

plt.subplot(1, 3, 1)
sns.histplot(df['Purchase Price'], kde=True, bins=10, color='blue')
plt.title('Distribusi Purchase Price')

plt.subplot(1, 3, 2)
sns.histplot(df['Current Price'], kde=True, bins=10, color='green')
plt.title('Distribusi Current Price')

plt.subplot(1, 3, 3)
sns.histplot(df['Gain/Loss'], kde=True, bins=10, color='red')
plt.title('Distribusi Gain/Loss')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data
df = pd.read_csv("/content/stock.csv")

# Clean and convert Dividend/Share (if necessary)
df["Dividend/share"] = df["Dividend/share"].str.lstrip("$").astype(float)

# Filter data dengan dividen 5 tertinggi
top_5_dividends = df.nlargest(5, "Dividend/share")

top_5_dividends["Symbol and Stock Name"] = top_5_dividends["Symbol"] + " - " + top_5_dividends["Stock Name"]

# Visualisasi menggunakan seaborn
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.barplot(x="Symbol and Stock Name", y="Dividend/share", data=top_5_dividends)
plt.xlabel("Symbol dan Stocok Name")
plt.ylabel("Dividend / Share")
plt.title("5 Stock Name dan Symbol dengan Dividen/Share 5 tertinggi")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data
df = pd.read_csv("/content/stock.csv")

# Menghapus tanda dolar
df["Market Value"] = df["Market Value"].str.replace("$", "")

# Menghapus tanda koma (pemisah ribuan) dan mengubah titik (pemisah desimal) menjadi koma
df["Market Value"] = df["Market Value"].str.replace(",", "").str.replace(".", "")

# Konversi string menjadi float
df["Market Value"] = df["Market Value"].astype(float)

# Filter 5 nilai Market Value tertinggi
top_5_market_value = df.nlargest(5, "Market Value")

top_5_market_value["Symbol and Stock Name"] = top_5_market_value["Symbol"] + " - " + top_5_market_value["Stock Name"]

# Visualisasi menggunakan seaborn
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.barplot(x="Symbol and Stock Name", y="Market Value", data=top_5_market_value)
plt.xlabel("Symbol dan Stocok Name")
plt.ylabel("Market Value")
plt.title("5 Stock Name dan Symbol dengan Market Value 5 tertinggi")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
